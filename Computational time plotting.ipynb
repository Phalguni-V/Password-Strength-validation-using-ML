```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the data for Dataset 1
data1 = {
    'Length': [8, 9, 10, 11, 12, 13, 14, 15, 16],
    'Logistic Regression Time (minutes)': [0.41, 0.27, 0.44, 0.34, 0.58, 0.61, 0.31, 0.33, 0.54],
    'Decision Tree Time (minutes)': [0.06, 0.05, 0.13, 0.11, 0.12, 0.08, 0.11, 0.04, 0.09],
    'Random Forest Time (minutes)': [1.57, 1.48, 2.83, 3.14, 2.60, 2.92, 2.67, 2.41, 3.30]
}

# Define the data for Dataset 2
data2 = {
    'Length': [8, 12, 16, 20, 24, 28, 32, 36, 40],
    'Logistic Regression Time (minutes)': [4.15, 6.37, 7.80, 37.21, 17.54, 24.87, 22.23, 29.07, 37.66],
    'Decision Tree Time (minutes)': [0.08, 0.16, 0.10, 0.10, 0.14, 0.20, 0.16, 0.13, 0.47],
    'Random Forest Time (minutes)': [8.54, 9.44, 7.26, 18.91, 14.83, 11.23, 7.85, 13.80, 11.00]
}
```

```python
# Convert to DataFrame
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# Plotting
plt.figure(figsize=(14, 5))

# Plot for Dataset 1
plt.subplot(1, 2, 1)
plt.plot(df1['Length'], df1['Logistic Regression Time (minutes)'], marker='o', color='coral', label='Logistic Regression')
plt.plot(df1['Length'], df1['Decision Tree Time (minutes)'], marker='o', color='#2ca02c', label='Decision Tree')
plt.plot(df1['Length'], df1['Random Forest Time (minutes)'], marker='o', color='#1f77b4', label='Random Forest')
plt.xlabel('Length')
plt.ylabel('Time (minutes)')
plt.title('Dataset 1')
plt.legend(title='Algorithm')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(14, 5))

# Plot for Dataset 2
plt.subplot(1, 2, 2)
plt.plot(df2['Length'], df2['Logistic Regression Time (minutes)'], marker='o', color='coral', label='Logistic Regression')
plt.plot(df2['Length'], df2['Decision Tree Time (minutes)'], marker='o', color='#2ca02c', label='Decision Tree')
plt.plot(df2['Length'], df2['Random Forest Time (minutes)'], marker='o', color='#1f77b4', label='Random Forest')
plt.xlabel('Length')
plt.ylabel('Time (minutes)')
plt.title('Dataset 2')
plt.legend(title='Algorithm')
plt.grid(True)

# Setting x-axis ticks to intervals of 4
plt.xticks(ticks=range(8, 41, 4))

plt.tight_layout()
plt.show()
```
